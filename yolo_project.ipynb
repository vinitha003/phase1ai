```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

```

```python
image = cv2.imread("/content/image.jpg")
```

```python
type(image)
```

```python
new_img = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(new_img)
```

```python
r,g,b = cv2.split(new_img)
print('r',r.shape)
print('g',g.shape)
print('b',b.shape)
new_img = cv2.merge((r,g,b))
```

```python
a = 12
w = int(new_img.shape[1]*a/100)
h = int(new_img.shape[0]*a/100)
dim = (w,h)
re_size = cv2.resize(new_img,dim,interpolation = cv2.INTER_AREA)
re_size.shape
```

```python
(h,w) = new_img.shape[:2]
c = (w/2,h/2)
angle = 90
m = cv2.getRotationMatrix2D( c,angle,1.0)
rotate_90 = cv2.warpAffine(new_img,m,(h,w))
```

```python
plt.imshow(rotate_90)
```
